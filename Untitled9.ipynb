```python
# -------------------------------
# 1️⃣ Upload CSV file
# -------------------------------
from google.colab import files
uploaded = files.upload()  # Upload "Mall_Customers.csv" from your computer

# -------------------------------
# 2️⃣ Import libraries & load data
# -------------------------------
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# Load dataset
df = pd.read_csv("Mall_Customers.csv")
df.head()

# -------------------------------
# 3️⃣ Prepare data (Annual Income & Spending Score)
# -------------------------------
X = df[['Annual Income (k$)', 'Spending Score (1-100)']].values

# Scale features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# -------------------------------
# 4️⃣ Elbow Method to find optimal K
# -------------------------------
wcss = []
for k in range(1, 11):
    km = KMeans(n_clusters=k, random_state=42)
    km.fit(X_scaled)
    wcss.append(km.inertia_)

plt.plot(range(1, 11), wcss, marker='o')
plt.title("Elbow Method to find Optimal K")
plt.xlabel("Number of clusters (K)")
plt.ylabel("WCSS")
plt.show()

# -------------------------------
# 5️⃣ Apply KMeans with optimal K
# -------------------------------
k_opt = 5  # You can adjust if Elbow suggests a different K
kmeans = KMeans(n_clusters=k_opt, random_state=42)
clusters = kmeans.fit_predict(X_scaled)  # ✅ This defines 'clusters'

# Add cluster info to dataframe (optional)
df['Cluster'] = clusters

# -------------------------------
# 6️⃣ Visualize clusters in 2D
# -------------------------------
plt.figure(figsize=(8,6))
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=clusters, cmap='viridis', s=50)
plt.xlabel("Annual Income (scaled)")
plt.ylabel("Spending Score (scaled)")
plt.title(f"K-Means Clusters (K={k_opt})")
plt.show()

# -------------------------------
# 7️⃣ Quick insights
# -------------------------------
print("Cluster counts:\n", df['Cluster'].value_counts())
```
